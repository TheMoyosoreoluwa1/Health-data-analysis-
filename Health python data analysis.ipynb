About this file
Here are the bullet point descriptions for each column name:
1 Entity: The name of the country or geographic region for which the data is reported.
2 Code: A three-letter country code assigned by the International Organization for Standardization (ISO).
3 Year: The year in which the data was collected or reported.
4 Deaths - Neoplasms - Sex: Both - Age: Under 5 (Rate): The number of deaths from neoplasms (i.e., cancer) per 100,000 people under the age of 5 in the given country or region.
5 Deaths - Neoplasms - Sex: Both - Age: Age-standardized (Rate): The number of deaths from neoplasms per 100,000 people, adjusted for differences in age distribution across populations.
6 Deaths - Neoplasms - Sex: Both - Age: All Ages (Rate): The number of deaths from neoplasms per 100,000 people across all age groups in the given country or region.
